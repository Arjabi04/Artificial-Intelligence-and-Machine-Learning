### **Task - 1:**

Create a Python program that converts between different units of measurement.

• The program should:
  1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
  2. Ask the user to input the value to be converted.
  3. Perform the conversion and display the result.
  4. Handle potential errors, such as invalid input or unsupported conversion types.

• Requirements:
1. **Functions**: Define at least one function to perform the conversion.
2. **Error Handling**: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. **User Input**: Prompt the user to select the conversion type and input the value.
4. **Docstrings**: Include a docstring in your function to describe its purpose, parameters, and

• Conversion Options:
1. **Length**:
– Convert meters (m) to feet (ft).
– Convert feet (ft) to meters (m).
2. **Weight**:
– Convert kilograms (kg) to pounds (lbs).
– Convert pounds (lbs) to kilograms (kg).
3. **Volume**:
– Convert liters (L) to gallons (gal).
– Convert gallons (gal) to liters (L).

```python
def convert_units(conversion_type, value):
    """
    Converts units based on the selected conversion type.

    Parameters:
    conversion_type (int): The type of conversion selected by the user.
    value (float): The numeric value to be converted.

    Returns:
    float: The converted value.
    """

    if conversion_type == 1:
        return value * 3.28084  # meters to feet
    elif conversion_type == 2:
        return value / 3.28084  # feet to meters
    elif conversion_type == 3:
        return value * 2.20462  # kg to pounds
    elif conversion_type == 4:
        return value / 2.20462  # pounds to kg
    elif conversion_type == 5:
        return value * 0.264172  # liters to gallons
    elif conversion_type == 6:
        return value / 0.264172  # gallons to liters
    else:
        return None


print("Unit Converter")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    choice = int(input("Select conversion type (1-6): "))
    value = float(input("Enter the value to convert: "))

    result = convert_units(choice, value)

    if result is None:
        print("Unsupported conversion type.")
    else:
        print("Converted value:", result)

except ValueError:
    print("Error: Please enter a valid numeric value.")
```

```python
def find_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The largest number in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The smallest number in the list.
    """
    return min(numbers)


print("Math Operations on a List")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

try:
    choice = int(input("Choose an operation (1-4): "))

    user_input = input("Enter numbers separated by spaces: ")

    numbers = [float(num) for num in user_input.split()]

    if len(numbers) == 0:
        raise ValueError("The list cannot be empty.")

    if choice == 1:
        print("Sum:", find_sum(numbers))
    elif choice == 2:
        print("Average:", find_average(numbers))
    elif choice == 3:
        print("Maximum:", find_maximum(numbers))
    elif choice == 4:
        print("Minimum:", find_minimum(numbers))
    else:
        print("Invalid operation selected.")

except ValueError:
    print("Error: Please enter valid numbers.")
```

### **1. Flatten a Nested List:**

Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

**• Requirements:**

– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.

– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
    lst (list): A nested list containing sublists.

    Returns:
    list: A single flattened list containing all elements.
    """
    flat_list = []

    for sublist in lst:
        for item in sublist:
            flat_list.append(item)

    return flat_list

numbers = [[1, 2], [3, 4], [5]]
result = flatten(numbers)

print(result)
```

### **2. Accessing Nested List Elements:**

Write a Python function that extracts a specific element from a nested list given its indices.

**• Requirements:**

– Define a function access nested element(lst, indices) that takes a nested list lst and
a list of indices indices, and returns the element at that position.

– Example: For the input lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] with indices = [1,
2], the output should be 6.

```python
def access_nested_element(lst, indices):
    """
    Accesses an element in a nested list using a list of indices.

    Parameters:
    lst (list): The nested list.
    indices (list): A list of indices to traverse the nested list.

    Returns:
    The element at the specified position.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)

print(result)
```

### **3. Sum of All Elements in a Nested List:**

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

**• Requirements:**

– Define a function sum nested(lst) that takes a nested list lst and returns the sum of all
the elements.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21.

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list, regardless of depth.

    Parameters:
    lst (list): The nested list containing numbers or sublists.

    Returns:
    int/float: The sum of all numbers in the nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # recursive call for sublist
        else:
            total += item
    return total

numbers = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(numbers)

print(result)
```

### **4. Remove Specific Element from a Nested List:**

Write a Python function that removes all occurrences of a specific element from a nested list.

**• Requirements:**

– Define a function remove element(lst, elem) that removes elem from lst and returns the
modified list.

– Example: For the input lst = [[1, 2], [3, 2], [4, 5]] and elem = 2, the output should
be [[1], [3], [4, 5]].

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): The nested list.
    elem : The element to remove.

    Returns:
    list: The modified list with the element removed.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            # Recursively remove elem from sublist
            result.append(remove_element(item, elem))
        else:
            if item != elem:
                result.append(item)
    return result


nested_list = [[1, 2], [3, 2], [4, 5]]
element_to_remove = 2

result = remove_element(nested_list, element_to_remove)
print(result)
```

### **5. Find the Maximum Element in a Nested List:**

Write a Python function that finds the maximum element in a nested list (regardless of depth).

**• Requirements:**

– Define a function find max(lst) that takes a nested list lst and returns the maximum
element.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list regardless of depth.

    Parameters:
    lst (list): The nested list containing numbers or sublists.

    Returns:
    int/float: The maximum number in the nested list.
    """
    max_val = None

    for element in lst:
        if isinstance(element, list):
            current_val = find_max(element)
        else:
            current_val = element

        # If current_val is None (empty sub-list), skip it
        if current_val is None:
            continue

        # If max_val hasn't been set yet, or we found a bigger number
        if max_val is None or current_val > max_val:
            max_val = current_val

    return max_val


numbers = [[1, 2], [3, [4, 5]], 6]
result = find_max(numbers)

print(result)
```

### **6. Count Occurrences of an Element in a Nested List:**

Write a Python function that counts how many times a specific element appears in a nested list.

**• Requirements:**

– Define a function count occurrences(lst, elem) that counts the occurrences of elem in
the nested list lst.

– Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should
be 3.

```python
def count_occurrences(lst, elem):
    """
    Counts the number of times a specific element appears in a nested list.

    Parameters:
    lst (list): The nested list.
    elem : The element to count.

    Returns:
    int: The number of occurrences of elem in the nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            # Recursively count in sublist
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count


# Example
nested_list = [[1, 2], [2, 3], [2, 4]]
element_to_count = 2

result = count_occurrences(nested_list, element_to_count)
print(result)
```

### **7. Flatten a List of Lists of Lists:**

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

**• Requirements:**

– Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single
flattened list.

– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8].



```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list regardless of depth.

    Parameters:
    lst (list): A nested list containing sublists at any depth.

    Returns:
    list: A single flattened list containing all elements.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # recursive call
        else:
            flat_list.append(item)
    return flat_list


# Example
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)

print(result)
```

### **8. Nested List Average:**

Write a Python function that calculates the average of all elements in a nested list.

**• Requirements:**

– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.

– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    """
    Calculates the average of all numbers in a nested list.

    Parameters:
    lst (list): A nested list containing numbers or sublists.

    Returns:
    float: The average of all numbers in the nested list.
    """
    def flatten_and_sum(lst):
        total = 0
        count = 0
        for item in lst:
            if isinstance(item, list):
                sub_total, sub_count = flatten_and_sum(item)
                total += sub_total
                count += sub_count
            else:
                total += item
                count += 1
        return total, count

    total_sum, total_count = flatten_and_sum(lst)
    if total_count == 0:
        return 0  # avoid division by zero
    return total_sum / total_count


nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)

print(result)
```


### 4. Matrix Multiplication

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.

```python
import time
import numpy as np

# Create two 1000x1000 matrices using lists
size = 1000
# Python lists
matrix1_list = [[i + j for j in range(size)] for i in range(size)]
matrix2_list = [[i - j for j in range(size)] for i in range(size)]

# NumPy arrays
matrix1_np = np.array(matrix1_list)
matrix2_np = np.array(matrix2_list)

# 1. Matrix multiplication using Python lists (nested loops)
start_time = time.time()

result_list = [[0]*size for _ in range(size)]
for i in range(size):
    for j in range(size):
        sum_val = 0
        for k in range(size):
            sum_val += matrix1_list[i][k] * matrix2_list[k][j]
        result_list[i][j] = sum_val

end_time = time.time()
list_time = end_time - start_time
print("Time taken using Python lists (1000x1000): {:.2f} seconds".format(list_time))

# 2. Matrix multiplication using NumPy
start_time = time.time()
result_np = np.dot(matrix1_np, matrix2_np)
end_time = time.time()
numpy_time = end_time - start_time
print("Time taken using NumPy arrays (1000x1000): {:.2f} seconds".format(numpy_time))
```
